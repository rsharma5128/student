## HomeWork Boolean
Write a function that checks if a number is positive AND even

Test it with an input of 8

Hint: use the modulo operator (%) to check if a number is even, and comparison operators to check if it's positive.


```python
%%js
// CODE_RUNNER: Create a function that checks if a number is positive and even

function isPositiveAndOdd(num) {
    let isPositive = num > 0;
    let isOdd = num % 2 !== 0;
    return isPositive && isOdd;
}

console.log(isPositiveAndOdd(8)); // false
console.log(isPositiveAndOdd(-3)); // false
console.log(isPositiveAndOdd(98))

function hasanevensquarerootAndisdivisiblebythree(num) {
    let hasanevensquareroot = Math.sqrt(num) % 2 === 0;
    let isdivisiblebythree = Math.sqrt(num) % 3 === 0;
    return hasanevensquareroot && isdivisiblebythree;
}
console.log(hasanevensquarerootAndisdivisiblebythree(36)); // true
console.log(hasanevensquarerootAndisdivisiblebythree(16)); // false
console.log(hasanevensquarerootAndisdivisiblebythree(18)); // false
```
